```python
df['rating'] = pd.qcut(
    df['fico_score'],
    q=10,
    labels=False
)
```

```python
bucket_bounds = pd.qcut(
    df['fico_score'],
    q=10,
    retbins=True
)[1]

print(bucket_bounds)
```

```python
def fico_to_rating(fico):

    for i in range(
        len(bucket_bounds)-1
    ):

        if (
            bucket_bounds[i]
            <= fico
            <= bucket_bounds[i+1]
        ):
            return 9-i

    return None
```

```python
print(
    fico_to_rating(800)
)

print(
    fico_to_rating(650)
)

print(
    fico_to_rating(500)
)
```
